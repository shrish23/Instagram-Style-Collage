```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
tl = cv.imread('images/top_left.jpg')
tr = cv.imread('images/top_right.jpg')
bl = cv.imread('images/bottom_left.jpg')
br = cv.imread('images/bottom_right.jpg')
ce = cv.imread('images/center.jpeg')
```

```python
#REsizing 
tl = cv.cvtColor(tl,cv.COLOR_BGR2RGB)
tl = cv.resize(tl,(200,200))

tr = cv.cvtColor(tr,cv.COLOR_BGR2RGB)
tr = cv.resize(tr,(200,200))

bl = cv.cvtColor(bl,cv.COLOR_BGR2RGB)
bl = cv.resize(bl,(200,200))

br = cv.cvtColor(br,cv.COLOR_BGR2RGB)
br = cv.resize(br,(200,200))

ce = cv.cvtColor(ce,cv.COLOR_BGR2RGB)
ce = cv.resize(ce,(100,100))
```

```python
collage = np.zeros((430,430,3),np.uint8)
plt.imshow(collage)
print(collage.shape)
```

```python
collage[10:210,10:210,:] = tl[:,:,:]
plt.imshow(collage)
print(collage.shape)
```

```python
collage[220:420,10:210,:] = bl[:,:,:]
plt.imshow(collage)
print(collage.shape)
```

```python
collage[220:420,220:420,:] = br[:,:,:]
plt.imshow(collage)
print(collage.shape)
```

```python
collage[10:210,220:420,:] = tr[:,:,:]
plt.imshow(collage)
print(collage.shape)
```

```python
ce = cv.copyMakeBorder(ce,10,10,10,10,cv.BORDER_CONSTANT)
```

```python
collage[155:275,155:275,:] = ce[:,:,:]
plt.imshow(collage)
plt.axis("off")
print(collage.shape)
```

```python
fin = cv.cvtColor(collage,cv.COLOR_BGR2RGB)
cv.imwrite("Final_Image.jpg",fin)
```
